### Ejemplo de MINI sistema de punto flotante (usando Jupyter Notebook)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[];
#Generar para todos los n√∫meros positivos del sistema (2,3,-2,3)
mantisa=np.arange(1.0, 1.8, 0.25)
for i in mantisa:
    for j in range(6):
        x.append(i*2**(j-2))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Define flatten to flatten a list of lists
import collections
def flatten(x):
    if isinstance(x, collections.Iterable):
        return [a for i in x for a in flatten(i)]
    else:
        return [x]
```

```python
#Agregar los negativos
vecm1=(-1)*np.ones(len(x))
posn=np.array([vecm1*x,x,0])
posn=np.array(list(flatten(posn)))
posn.sort()
y=np.zeros(len(posn))
plt.figure(figsize=(20,10))
plt.plot(posn,y,'+',markersize=15)
plt.title('Sistema de punto flotante (2,3,-2,3)')
plt.grid(True)
```
