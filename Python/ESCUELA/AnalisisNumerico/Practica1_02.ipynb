## Ejercicio 1
Escribe un  programa que calcule los errores absoluto y relativo de la aproximación de Stirling

$$n! \approx \sqrt{2\pi n}\left(\frac{n}{e} \right)^n,$$

para $n=1,2,...,10.$

```python
#Cálculo de el factorial de un número natural
from math import pi,e,sqrt
n = int(input('Introduce un número natural, n = ', ))

k = 1
for i in range(n):
    k = k*(i+1)       # Primero calculemos el factorial de forma iterativa usando un bucle for
    fact = sqrt(2*pi*(i+1))*(((i+1)/e)**(i+1)) # Calculamos el factorial con Stirling
    print('El factorial de', i+1, 'usando el bucle for es:', k)
    print('El factorial de', i+1, 'usando la aproximacion de Stirling es:', fact)
    errorAbs = k-fact
    print('Error absoluto:', errorAbs)
    print('Error relativo:', (errorAbs/k))
    print()
```

## Ejercicio 4
Escribe un programa que, mediante el método de eliminación de Gauss-Jordan y sin pivoteo, calcule la inversa de una matriz de $n \times n$. Toma en cuenta las restricciones y condiciones que se deben cumplir para poder utilizar este método.

```python
import numpy as np 
import sys

def inversa(M):
  n = len(M)
  A = np.zeros((n,2*n))

  #Aumentamos M con A
  for i in range(n):
    for j in range(n):
      A[i][j] = M[i][j]

  #Idedntidad
  for i in range(n):        
      for j in range(n):
          if i == j:
              A[i][j+n] = 1

  #GAuss Jordan
  for i in range(n):
      if A[i][i] == 0.0:
          sys.exit('Division entre cero')
      for j in range(n):
          if i != j:
              ratio = A[j][i]/A[i][i]
              for k in range(2*n):
                  A[j][k] = A[j][k] - ratio * A[i][k]
  for i in range(n):
      divisor = A[i][i]
      for j in range(2*n):
          A[i][j] = A[i][j]/divisor

  return A
```

Escribe un programa que calcule la norma infinito o la norma 1 de una matriz de $n\times n$. Luego, utiliza este programa y el anterior para calcular el número de condición de una matriz de $n \times n$ sin utilizar las funciones predeterminadas en Python. Prueba con las matrices:

$$A = \left(\begin{array}{cc} 1.0125 & 2 & 3\\ 1 & 2 & 3  \\ 0 & 0 & 1\end{array}\right) \quad\textrm{y}\quad B = \left(\begin{array}{cc} -2 & 1 & 1\\ -2 & -5 & 1 \\ 1 & 0 & 2\end{array}\right)$$.

```python
from math import sqrt 
import numpy as np

def normaFrobenius(M):
    n = len(M)
    suma = 0.0
    for i in range(n):
        for j in range(n):
            suma += M[i][j] ** 2
    res = sqrt(suma)
    return round(res, 5)

def cond(M):
  return normaFrobenius(M) * normaFrobenius(inversa(M))
  
A = np.array([[1.0125, 2.0 ,3.0], [1.0,2.0,3.0], [0.0,0.0,1.0]])
B = np.array([[-2, 1 ,1], [-2,-5,1], [1,0,2]])

print("Matriz A:", A)
print("Número de condición: " ,  cond(A))
print("Matriz B:", B)
print("Número de condición: " ,  cond(B))
```

## Ejercicio 5

La matriz de Hilbert de $n\times n$ se define como aquella que tiene entradas $\left(\rm{H_n}\right)_{ij} = \frac{1}{i+j-1}$ para $i,j=1,2,\cdots,n$. Escribe un programa que genere la matriz de Hilbert $\rm{H_n}$ de $n\times n$, y que construya una tabla que muestre $\rm{Cond}(H_n)$ y $\rm{Det(H_n)}\times\rm{Det(H^{-1}_n)}$ para $n = 1,2,...,30$.

```python
from tabulate import tabulate
import numpy as np

def Hilbert(n):
  return [[1/(i+j-1) for j in range(1,n+1)] for i in range(1,n+1)]

def ejer_4(n):
  if n == 0:
    return [0,0,0,0]
  H = Hilbert(n)
  c = cond(H)
  detH = np.linalg.det(H)
  det_H = np.linalg.det(np.linalg.inv(H))
  return [n,c,detH,det_H]

tabla = [ejer_4(i) for i in range(0,31)]
tabla[0] = ['n', 'Cond(H_n)', 'det(H_n)', 'det(H_n^(-1))']
print(tabulate(tabla, headers='firstrow', tablefmt='fancy_grid'))
```

## Ejercicio 6
Construye tres rutinas que resuelvan sistemas de ecuaciones lineales mediante los métodos: eliminación gaussiana, factorización LU y factorización de Cholesky. Toma en consideración las restricciones y consideraciones que se deben cumplir para cada método.

```python
# Se importan las bibliotecas necesarias
import numpy as np

def pivote(M):
    """Regresa el pivote de la matriz M, usado en el metodo de Doolittle."""
    m = len(M)                                                                                                                                                                                       
    id_mat = np.identity(m)
                                                                                                                                                                                               
    for j in range(m):
        row = max(range(j, m), key=lambda i: abs(M[i][j]))
        if j != row:                                                                                                                                                                                                                           
            id_mat[j], id_mat[row] = id_mat[row], id_mat[j]

    return id_mat

def LU(A):
    """Asumimos que A es cuadrada"""
    n = len(A)
                                                                                                                                                                                                            
    L = np.zeros(n)
    U = np.zeros(n)

    # Pivote P y PA                                                                                                                                                                                          
    P = pivote(A)
    PA = np.matmul(P, A)

    # Decomposicion                                                                                                                                                                                                                    
    for j in range(n):
        # Las diagonales de L serán la unidad                                                                                                                                                                                                 
        L[j][j] = 1.0

        # u_{ij} = a_{ij} - \sum_{k=1}^{i-1} u_{kj} l_{ik}                                                                                                                                                                                      
        for i in range(j+1):
            s1 = sum(U[k][j] * L[i][k] for k in range(i))
            U[i][j] = PA[i][j] - s1

        # LaTeX: l_{ij} = \frac{1}{u_{jj}} (a_{ij} - \sum_{k=1}^{j-1} u_{kj} l_{ik} )                                                                                                                                                                  
        for i in range(j, n):
            s2 = sum(U[k][j] * L[i][k] for k in range(j))
            L[i][j] = (PA[i][j] - s2) / U[j][j]

    return (P, L, U)

def Choleski(A):
  #Construcción de la matriz L
  n = len(A)
  L = np.array([[0.0 for j in range(n)] for i in range(n)]) #Matriz de ceros, aquí alojaremos los valores de L
  for i in range(n):
      for j in range(n):           #Construcción de los elementos que están debajo de la diagonal de L
          if (j < i):              #Esta condición es importante porque solo considera elementos debajo de la diagonal de L
              S = 0.0
              for k in range(j):
                  S = S + L[i,k]*L[j,k]
              L[i,j] = (A[i,j] - S)/L[j,j]       
      SD = 0.0
      for k in range(i):          #Construcción de los elementos de la diagonal de L
          SD = SD + (L[i,k])**2  
      L[i,i] = sqrt(A[i,i] - SD)
```
